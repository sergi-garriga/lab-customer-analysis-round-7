## Now we will try to check the normality of the numerical variables visually
## Use seaborn library to construct distribution plots for the numerical variables.

```python
# We use a for loop to iterate over all the numerical columns.

for col in numerical.columns:
    sns.displot(x=numerical[col], kde=True, height=5, aspect=15/5)
    plt.show()
```

## Use Matplotlib to construct histograms

```python
for col in numerical.columns:
    f, ax = plt.subplots(figsize=(15, 5))
    plt.hist(numerical[col])
    plt.xlabel(col)
    plt.show()
```

```python
# In Matplotlib, the number of bins is not related to the number of entries for each column.
# We will set the number of bins acoording to the range of each feature.
# Add some settings for the output.

for col in numerical.columns:
    col_range = np.percentile(numerical[col],100)-np.percentile(numerical[col],0) 
    if (col_range < 100):
        number_of_bins = 10
    elif (col_range < 1000):
        number_of_bins = (col_range/10)
    elif (col_range < 10000):
        number_of_bins = (col_range/100)
    elif (col_range < 100000):
        number_of_bins = (col_range/1000)
    else:
        number_of_bins = 100
    f, ax = plt.subplots(figsize=(15, 5))
    plt.hist(numerical[col], bins = int(number_of_bins), color = "goldenrod", lw=0.5, ec="snow", alpha=0.65, density=False)
    plt.xlabel(col.upper(), fontweight="bold", fontsize=10)
    plt.show()
```

# Do the distributions for different numerical variables look like a normal distribution

```python
# Each feature might need a different approach.

# For the Customer Lifetime Value, we will simply apply the boxcox transformation. 

# Wea apply the Boxcox but don't overwrite the original data, so it does not interfere with other operations.
original_data_costumer_lifetime_value = numerical["Customer Lifetime Value"]
fitted_data_costumer_lifetime_value = boxcox(original_data_costumer_lifetime_value, 0)
 
# Subplots to compare the original and the result
fig, ax = plt.subplots(1, 2)
 
# Call the plots
sns.histplot(original_data_costumer_lifetime_value, kde = True,
            label = "Original Data", ax = ax[0])
 
sns.histplot(fitted_data_costumer_lifetime_value, kde = True,
            label = "Fitted Data", ax = ax[1])
 
# Adding legends to the subplots
plt.legend(loc = "upper right")
 
# rescaling the subplots
fig.set_figheight(5)
fig.set_figwidth(15)
 
```

```python
# For the Income feature, we will drop the zeros, because they are not significant distribution of income values.
# Unnemployed people might be taken into consideration in other 
# We will the apply the Boxcox transformation.

# Make a new dataframe only taking the non-zero values for the Income feature.
income_col_cleaned = numerical["Income"].loc[(numerical["Income"] != 0)]

# apply the boxcox to the Income feature
income_col_cleaned_boxcox = boxcox(income_col_cleaned, 0)

# Subplots to compare the original and the result

plt.figure(figsize = [15, 5])

# Subplot original.
plt.subplot(1, 3, 1) # 1 row, 3 cols, subplot 1
plt.hist(x=numerical["Income"], bins = 100, color = "goldenrod", lw=0.5, ec="snow", alpha=0.65, density=False)

# Subplot cleaned of zeros
plt.subplot(1, 3, 2) # 1 row, 3 cols, subplot 2
plt.hist(x=income_col_cleaned, bins = 100, color = "goldenrod", lw=0.5, ec="snow", alpha=0.65, density=False)

# Subplot cleaned of zeros and boxcoxed.
plt.subplot(1, 3, 3) # 1 row, 3 cols, subplot 3
plt.hist(x=income_col_cleaned_boxcox, bins = 100, color = "goldenrod", lw=0.5, ec="snow", alpha=0.65, density=False)

plt.show()
```
